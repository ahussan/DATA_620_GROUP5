To begin with this assignment, let's import Networkx and matplotlib

```python
import networkx as nx
import matplotlib.pyplot as plt
```

Create a new instance of the NetworkX Graph

```python
G = nx.Graph()
```

Add edge as provoded in the assignment description

```python
G.add_edge("Andre", "Carol")
G.add_edge("Andre", "Beverly")
G.add_edge("Andre", "Diane")
G.add_edge("Andre", "Fernando")
G.add_edge("Carol", "Fernando")
G.add_edge("Carol", "Diane")
G.add_edge("Fernando", "Garth")
G.add_edge("Fernando", "Heather")
G.add_edge("Fernando", "Diane")
G.add_edge("Garth", "Beverly")
G.add_edge("Garth", "Diane")
G.add_edge("Garth", "Ed")
G.add_edge("Garth", "Heather")
G.add_edge("Heather", "Ike")
G.add_edge("Jane", "Ike")
G.add_edge("Diane", "Ed")
G.add_edge("Diane", "Beverly")
G.add_edge("Beverly", "Ed")
```

Now that all of our edges are added into the graph, we can draw the graph. Notice that the nodes of the graph automatically defined as we added the edges. 

```python
 nx.draw(G, with_labels = True, node_size=800, node_color='c')
```

The above graphs is perfect for initial attempt. however, each time we run the above code to generage a map, the shape of the graph and the node position will be changed. Therefore we need to specify the node position using the attribute called pos

```python
pos={'Andre':(0,1),'Diane':(1,0),
     'Heather':(3,0),'Ike':(4,0),
     'Jane':(5,0),'Fernando':(2,1),
     'Carol': (1, 2), 'Beverly': (0, -1),
     'Garth': (2, -1), 'Ed':(1, -2)}    
nx.set_node_attributes(G, pos, 'coord')
```

The string 'coord' indicates that we are passing coordinates of the nodes in the pos as a dictionary. 

Now that we have set the pos of our graph, we can 

```python
nx.draw(G, pos, with_labels = True, node_size=800, node_color='c')
```
